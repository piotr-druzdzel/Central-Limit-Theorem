# Demonstration of the central limit theorem

```python
# imports
from numpy.random import seed
from numpy.random import randint
from numpy import mean

from matplotlib import pyplot
import seaborn as sns

import statistics
```

```python
# seed the random number generator
seed(1)
```

## 30 rolls

```python
rolls = randint(1, 7, 30)
rolls
```

```python
sns.distplot(rolls)
```

## 3000 rolls

```python
rolls = randint(1, 7, 3000)
rolls
```

```python
sns.distplot(rolls)
```

## Calculate the mean of 30 dice rolls 10000 times

```python
means = [mean(randint(1, 7, 30)) for _ in range(10000)]
```

```python
# plot the distribution of sample means
pyplot.hist(means)
pyplot.show()
```

## Standard deviation:

```python
statistics.stdev(means)
```
